```python
from fastbook import *
from fastai.vision.widgets import *

# python-dotenv
import os
from dotenv import load_dotenv
load_dotenv()   
```

```python
api_key = os.environ.get("AZURE_SEARCH_KEY") # put your Azure secret key as a env variable
```

```python
search_images_bing # our data source is this function provided by fastbook that will download images from bing for us
```

```python
results = search_images_bing(api_key, "front yard")
imgs = results.attrgot('content_url')
print(len(imgs))
```

```python
yard_types = ["front", "back"]
dl_path = Path("yard_imgs")

dl_path.mkdir(exist_ok = True)
for yard_type in yard_types:
    dest = (dl_path/yard_type)
    dest.mkdir(exist_ok = True)
    results = search_images_bing(api_key, f'{yard_type} yard')
    download_images(dest, urls = results.attrgot('contentUrl'))
```

```python
fns = get_image_files(dl_path)
f'{len(fns)} images'
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink) # remove failed imgs
```

## Load Data to Fast.ai 

```python
yards = DataBlock(
    blocks = (ImageBlock, CategoryBlock),
    get_items = get_image_files,
    splitter = RandomSplitter(valid_pct = 0.2, seed = 42),
    get_y = parent_label,
    item_tfms = Resize(128))
```

```python
dls = yards.dataloaders(dl_path)
```

```python
dls.valid.show_batch(max_n = 4, nrows = 1)
```

```python
# image sizes are not standardized
# to not squish/stretch the images (which makes the real-world out-of-domain!), we select a random resized crop each time
rrc_yards = yards.new(item_tfms = RandomResizedCrop(128, min_scale = 0.3))
rrc_dls = rrc_yards.dataloaders(dl_path)
rrc_dls.train.show_batch(max_n = 4, nrows = 1, unique = True) # show different crops of the same image
```

```python
# data augmentation example with fastbook
aug_yards = yards.new(item_tfms = Resize(128), batch_tfms = aug_transforms(mult = 2))
aug_dls = yards.dataloaders(dl_path)
aug_dls.train.show_batch(max_n = 8, nrows = 2, unique = True)
```

```python
# final data, includes rrc & aug
yards = yards.new(
    item_tfms = RandomResizedCrop(224, min_scale = 0.5),
    batch_tfms = aug_transforms())
dls = yards.dataloaders(dl_path)
```

```python
# train the data!
learner = cnn_learner(dls, resnet18, metrics = error_rate)
learner.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learner)
interp.plot_confusion_matrix()

# makes sense, since back yards usually more varied than front yards!
```
